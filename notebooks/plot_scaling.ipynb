Code to plot how our 3D cube prior scales.

```python
import plotly.graph_objects as go
import numpy as np

scene_scale = np.linspace(1, 1000, 1000)
size = 0.1
num_cubes = scene_scale**3 / size**3

# Create line graph
fig = go.Figure(data=go.Scatter(x=scene_scale, y=num_cubes, mode="lines"))

# Set chart title and axis labels
fig.update_layout(
    title="Number of cubes needed to fill a scene",
    xaxis_title="Scene scale",
    yaxis_title="Number of cubes"
)

# make pretty
font_size = 30
fig.update_layout(
    title_font_family="Times New Roman",
    title_font_color="black",
    title_font_size=font_size,
    legend_title_font_color="black",
    legend_font_size=font_size,
    legend_font_family="Times New Roman",
    xaxis=dict(
        title_font_family="Times New Roman",
        title_font_color="black",
        title_font_size=font_size,
        tickfont_family="Times New Roman",
        tickfont_color="black",
        tickfont_size=font_size,
    ),
    yaxis=dict(
        title_font_family="Times New Roman",
        title_font_color="black",
        title_font_size=font_size,
        tickfont_family="Times New Roman",
        tickfont_color="black",
        tickfont_size=font_size,
    )
)

# Display the chart
fig.show()

# save as PDF
fig.write_image("plot_scaling.pdf")
```
